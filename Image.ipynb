# New Section

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np

fig,ax = plt.subplots(figsize=(6,6))

head= plt.Circle((50, 60),7,color='cyan',fill=True)
ax.add_artist(head) #head

#eyes
eye1 = plt.Circle((46, 60),1,color='black',fill=True)
ax.add_artist(eye1)
eye2 = plt.Circle((54, 60),1,color='black',fill=True)
ax.add_artist(eye2)

#body
body = plt.Line2D([50,50], [53,20], color='black',linewidth=3)
ax.add_line(body)

#hands
lh = plt.Line2D([50,30], [40,30], color='black',linewidth=3)
rh = plt.Line2D([50,70], [40,30], color='black',linewidth=3)
ax.add_line(lh)
ax.add_line(rh)

#legs
ll = plt.Line2D([50,40], [20,0],color='black',linewidth=3)
rl = plt.Line2D([50,60], [20,0],color='black',linewidth=3)
ax.add_line(ll)
ax.add_line(rl)

#sun
sun = plt.Circle((80, 80), 6, color='yellow', fill=True)
ax.add_artist(sun)
sun_border= patches.Circle((80,80), 6, edgecolor='black', fill=False, linestyle='--', linewidth=2)
ax.add_artist(sun_border)

#set axes
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_yticks([0, 20, 40, 60, 80, 100])

plt.grid(linestyle='--')
plt.show()
```
