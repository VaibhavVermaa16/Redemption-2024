```python
df = pd.read_csv("matches.csv")
df.head(10)
```

```python
#Season which had the most number of matches
season = df.groupby("season")
season
```

```python
for year,data in season:
  print(year)
  print(data)
  print("\n")
```

```python
season_size = season.size()
```

```python
max_season_size = season_size.max()
max_season_size
```

```python
season_with_most_matches = season_size.idxmax()
season_with_most_matches
```

```python
print(season_with_most_matches," is the season with the most number of matches:", max_season_size)
```
