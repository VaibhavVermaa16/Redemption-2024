```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 =  pd.read_csv('original_data.csv')
df2 = pd.read_csv('synthetic_data.csv')
```

```python
df = pd.concat([df1,df2])
```

```python
#using boxplot
df.boxplot(figsize=(10,5))
plt.xticks(rotation=45)
plt.ylim(-5,5)
plt.show()
```

```python
#OUTLIEIR DETECTION USING IQR METHOD, used scatterplotss for eachfeature
Q1 =  df.quantile(0.25)
Q3 =df.quantile(0.75)
iqr =  Q3 - Q1

iqrr =(df <(Q1 - 1.5*iqr)) | (df > (Q3 +1.5*iqr))
for column in df.columns[:-1]:
    plt.figure(figsize=(6, 3))
    plt.scatter(df.index, df[column], c=iqrr[column])
    plt.ylabel(column)
    plt.xlabel('index')
    plt.title(f'{column}')
plt.tight_layout()
plt.show()
```

```python
# using z_score
df_zscore = (df -df.mean()) /df.std()

a=  df_zscore.shape[1]
plt.figure(figsize=(20, 30))
for i in range(1,a):
    plt.subplot(10,5,i)
    plt.scatter(df.index,df_zscore.iloc[:, i])
    plt.xlabel('indx')
    plt.title(df_zscore.columns[i])
    plt.ylabel('zscoreo')
plt.tight_layout()
plt.show()
```
